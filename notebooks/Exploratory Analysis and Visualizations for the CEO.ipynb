```python
import requests
import io
resp = requests.get(path).content
retail = pd.read_excel(resp)
```

```python
retail.head()
```

<b>Before I atempt answering any question, I consider it good practice for me to create a column for revenue, which would represent the product of the quantity and unit price for each entry. Besides revenue makes a large portion of my analysis.<b/>

```python
#Creating a column for revenue
retail['Revenue']  = retail['Quantity'] * retail['UnitPrice']
retail.Revenue = retail.Revenue.round(2)
```

```python
retail.head()
```

<b>Before commencing exploration i.e., answering the questions listed above. I would like to run a quick overview analysis on the revenue generated by each country in order for me to have a clear picture of the revenue situation of the business.

<b> In order to get a clear picture of the business revenue by country, I'd be grouping the data set by country.

```python
#Grouping the dataset by country and printing each country's share revenue.
country_revenue = retail.groupby('Country', as_index=False).sum().sort_values(by='Revenue', ascending=False)
country_revenue
```

```python
#Plotting the revenue share by country.
country_revenue.plot(x='Country', y='Revenue', kind='bar', color='b', figsize=(18,18))
plt.title('Revenue Share by Country', fontsize=18)
plt.xlabel('Country', fontsize=16)
plt.ylabel('Revenue (Dollar)', fontsize=16)
plt.show()
```

>## Deductions
><b>From the graph produced and the output displayed for the "Revenue Share by Country" above, it can be observed that;
> * <b>The United Kingdom has the highest Revenue share of ***\\$7, 308, 391.55***.
> * <b>While country that comes second in revenue share is the Netherlands with ***\\$285, 446.34***.
> * <b>Saudi Arabia contributes the least to the total revenue generated by the business with a contribution of ***\\$145.92***.
>
> <b>However, because of the huge difference in revenue generated between the United Kingdom and the Netherlands of ***\\$7, 022, 945.21***. I consider it good practice for my analysis to treat the United Kingdom as an outlier by exempting it from this current analysis, so as to capture a clear detail of the revenue generated by other countries in the dataset, so as to enable the business discover other areas of operation that has expansion opportunities.<b/>

```python
#Exempting the revenue share of the United Kingdom from the plot of Revenue Share by Country
country_revenue_excluding_UK = country_revenue.drop([35], axis=0)
country_revenue_excluding_UK.plot(x='Country', y='Revenue', kind='barh', color='b', figsize=(18,18))
plt.title('Revenue Share by Country', fontsize=18)
plt.xlabel('Country', fontsize=18)
plt.ylabel('Revenue (Dollar)', fontsize=18)
plt.show()
```

<b>Now that I have created a revenue by country view of the business, I would now go ahead with the intended the exploration.

 ## Question 1
<b>What is the time series of the revenue for the business for the 2011 buisness operation year only?

<b> I would like to begin this query by obtaining certain time based informations off the data set as shown below.

```python
#obtaining the minimum and maximum date time in the invoice date range
print(f"The first date on record in the data set is {retail['InvoiceDate'].min()}.")
print(f"The last date on record in the data set is {retail['InvoiceDate'].max()}.")
print(f"The range on the date record in the data set is {retail['InvoiceDate'].max() - retail['InvoiceDate'].min()}.")
```

> ## Deduction
><b>From the output above, I deduce that;
> * <b> The first date recorded in the dataset is ***2010-12-01***.
> * <b> The last date on record in the data set is ***2011-12-09***.
> * <b> The record spans for a period of ***373*** days.
>
> <b>However, to have a clear view of the monthly trend in revenue it would be good for analysis that the days in the year ***2010***, which happens to only be the days in the month of December ***2010*** be excluded from the record such that it shows only records for the year ***2011***. Moreover, in this respect it does appears that the CEO is only interest in insights that can be obtained from the year ***2011***.<b/>

```python
#Exempting date earlier that 2011-01-01 from the data set.
retail['InvoiceDate'] = pd.to_datetime(retail['InvoiceDate'])
date_excluding_2010 = retail[~(retail['InvoiceDate'] < '2011-01-01')]
#Checking the new date range in the record.
print(f"The first date on record in the data set is {date_excluding_2010['InvoiceDate'].min()}.")
print(f"The last date on record in the data set is {date_excluding_2010['InvoiceDate'].max()}.")
print(f"The range on the date record in the data set is {date_excluding_2010['InvoiceDate'].max() - date_excluding_2010['InvoiceDate'].min()}.")
```

<b>Haven successfully excluded the 2010 records off the dataset, I now have the above printed time based information about the dataset and can go ahead to run a monthly revenue trend on the dataset.<b/>

```python
#Converting the month numeral to name.
date_excluding_2010['InvoiceDate'] = date_excluding_2010['InvoiceDate'].dt.month_name().str[:3]
```

```python
#Grouping the dataset by date so as to reflect the monthly revenue.
date_excluding_2010 = date_excluding_2010.groupby(date_excluding_2010.InvoiceDate, as_index=False).sum().sort_values(by='InvoiceDate', ascending=True)
date_excluding_2010
```

```python
#Exempting the revenue share of the United Kingdom from the plot of Revenue Share by Country
date_excluding_2010.plot(x='InvoiceDate', y='Revenue', kind='bar', color='b', figsize=(18,18))
plt.title('Revenue Monthly Trend', fontsize=18)
plt.xlabel('Month', fontsize=18)
plt.ylabel('Revenue (Million Dollar)', fontsize=18)
plt.show()
```

```python
#Producing a line chart to get a trend.
plt.figure(figsize=(12, 6))
plt.plot(date_excluding_2010.InvoiceDate, date_excluding_2010.Revenue, 'o--b')
plt.xlabel('Month', fontsize=16)
plt.ylabel('Revenue (Dollar)', fontsize=16)
plt.title("Revenue Monthly Trend",  fontsize=18)
```

> ## Deduction
><b> From the outputs in this question, the revenue by month trend can be tracked from the line chart displayed above. It can therefore be observed that the month of November had the highest revenue generated with a total of ***\\$1, 161, 817.38*** generated. While the month that generated the least revenue is February, where only ***\\$447, 137.35*** was generated. The seemingly low revenue generated in December can be considered to be due to the fact that sales operations in December 2011 only lasted for 9 days, as sales records in the datset ended on 9th of December.<b/>

 ## Question 2
<b>What is the demand for the business products accross the regions where the business is operational?

<b>To obtain insights about the demands of the business products accross the regions where the business is operational, I'd be using the revenue by country dataset already developed ealier as it is also required that the original dataset be grouped by country. In this analysis I'd focus on the quantity feature of the dataset, as the quantity of the business products purchased can be considered a reflection of demand nature of the business.

```python
demand = country_revenue
demand
```

```python
#Plotting a barchart graph to reflect the business demand by country .
demand.plot(x='Country', y='Quantity', kind='bar', color='b', figsize=(18,18))
plt.title('Demand Share by Country', fontsize=18)
plt.xlabel('Country', fontsize=16)
plt.ylabel('Demand (Quantity)', fontsize=16)
plt.show()
```

> ## Deduction
><b>As in the case of the of the revenue by country analysis, we also have the demand of the United Kingdom leading that of other countries in the business with a very large margin, such that for the sake of good analytics, we can choose to view the demand data of the United Kingdom as an outlier, and focus on the demand of the other countries where the buisness can easily spot business expansion opportunities emanating from the analysis of their corresponding demands.

<b>I'd now be excluding the demand of the united Kingdom off the data set.

```python
#Exempting the demand data of the United Kingdom from the plot of demand Share by Country
country_revenue_excluding_UK = country_revenue.drop([35], axis=0)
country_revenue_excluding_UK.plot(x='Country', y='Quantity', kind='bar', color='b', figsize=(18,18))
plt.title('Demand Share by Country', fontsize=18)
plt.xlabel('Country', fontsize=18)
plt.ylabel('Demand (Quantity)', fontsize=18)
plt.show()
```

>## Deduction
><b>After excluding the United Kingdom from the data set, the following deductions can be made;
> * <b>Netherlands have the highest demand, with a demand of 200,361 and followed closely by Ireland with a total demand of 140,275.   
> * <b>The country with the least demand is Saudi Arabia, with the business products demand of 80.

<b>As a side analysis, I'd be attempting to draw a relationship between the revenue generated by each country and their respective business product demand.

<b>I'd be using bar chart to produce this representation, but I'd be focussing on the top 10 countries by revenue excluding the United Kingdom. But I would first atempt to obtain the view for the top 10 countries in terms of revenue and demand, so as to see if both categories contain exactly the same set of countries. As this would help for a clearer show of the relationship between their respective countries.

```python
#Obtaining the data for the top 10 revenues by country
revenue_excludingUK = country_revenue.drop([35], axis=0)
revenue_top10 = revenue_excludingUK.nlargest(10, 'Revenue')
revenue_top10
```

```python
#Obtaining a view for the top 10 demands by country.
demand_excludingUK = demand.drop([35], axis=0)
demand_top10 = demand_excludingUK.nlargest(10, 'Quantity')
demand_top10
```

> <b>Compairing the country feature for the two outputs above for revenue and demand, they both contain he same sets of countries. I can thus go ahead to run a representative visual of ther relationship.

```python
#Ploting the relationship between the demand and revenue of the countries with the highest demands and revenue share using 
#multiple(2) bar chart.

x = np.arange(len(demand_top10.Country))
width = .6 #the width of the bar
fig, ax = plt.subplots(figsize=(18, 18))
bar1 = ax.bar(x - width/2, demand_top10.Quantity, width, label='Demand')
bar2 = ax.bar(x + width/2, demand_top10.Revenue, width, label='Revenue')

ax.set_title('The Relationship between the Demand and Revenue of the top 10 countries', fontsize=18)
ax.set_xlabel('Country', fontsize=16)
ax.set_xticks(x, demand_top10.Country)
ax.legend()

ax.bar_label(bar1, padding=3)
ax.bar_label(bar2, padding=3)

plt.show()
```

> <b>Above is a visual represetation of the relationship between the respective demand and revenue of the top 10 countries where the business is operational.
>
><b>Taking a careful look at the graph, we can almost conclude that all the counrties represented above have some sort of relationship existing between their demands and the respective revenue they generate for the business.
